```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Constant Price Curve

BASE_ETH = 0.25

age = np.arange(0, 1500, 1)
price = np.full(1500, BASE_ETH)

plt.plot(age, price);
plt.xlabel("Output Noun Age")
plt.ylabel("Price (ETH)")
plt.title(f"Constant Price Curve \nbase (ETH) = {BASE_ETH}")
plt.savefig("img/constant-price-curve.png", facecolor=(1, 1, 1))
plt.show()
```

```python
# Output NFT age based price curve

BASE_ETH = 0.25
SLOPE_ETH_PER_AGE = 0.01

age = np.arange(0, 1500, 1)
price = BASE_ETH + SLOPE_ETH_PER_AGE * age;

plt.plot(age, price);
plt.xlabel("Output Noun Age")
plt.ylabel("Price (ETH)")
plt.title(f"Output NFT age based price curve \nbase (ETH) = {BASE_ETH}, slope (ETH/age) = {SLOPE_ETH_PER_AGE}")
plt.savefig("img/output-age-based-price-curve", facecolor=(1, 1, 1))
plt.show()
```

```python
# Age difference based price curve

BASE_ETH = 0.2
SLOPE_ETH_PER_AGE = 0.001

ageDifference = np.arange(-1500, 1500, 1)
price = BASE_ETH + SLOPE_ETH_PER_AGE * np.maximum(ageDifference, 0);

plt.plot(ageDifference, price);
plt.xlabel("AgeDiff = inputTokenId - outputTokenId")
plt.ylabel("Price (ETH)")
plt.title(f"Age difference based price curve \nbase (ETH) = {BASE_ETH}, slope (ETH/age) = {SLOPE_ETH_PER_AGE}")
plt.savefig("img/age-difference-price-curve.png", facecolor=(1, 1, 1))
plt.show()
```

```python
def plot_age_difference_price_curve(base_eth, slope_eth_per_age, age_range, ax, type):
    age_difference = np.arange(-age_range, age_range, 1)
    price = base_eth + slope_eth_per_age * np.maximum(age_difference, 0)
    ax.plot(age_difference, price)
    ax.set_ylabel("Price (ETH)")
    ax.set_title(f"{type}: base (ETH) = {base_eth}, slope (ETH/age) = {slope_eth_per_age}")

# Parameters
params = [
    (0.1, 0.01, "aggressive"),
    (0.2, 0.002, "moderate"),
    (0.3, 0.00, "fixed")
]
age_range = 1500

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 15), sharex=True)

# Plot for each set of parameters
for i, (base_eth, slope_eth_per_age, type) in enumerate(params):
    plot_age_difference_price_curve(base_eth, slope_eth_per_age, age_range, axs[i], type)

# Set common X label
axs[2].set_xlabel("ageDiff = inputTokenId - outputTokenId")

fig.suptitle("Age Difference Based Price Curves", fontsize=16)
# Adjust layout with vertical gaps
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.subplots_adjust(top=0.92, hspace=0.2)  # Adjust the height space between plots and add gap after title


# Adjust layout and save the figure
plt.savefig("img/age-difference-price-curve.png", facecolor=(1, 1, 1))
plt.show()
```
